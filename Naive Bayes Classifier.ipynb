# <font color='dark green'>Naive Bayes Classifier</font>

* Naive Bayes Classifier is based on <b>"Bayes" theorem</b> which gives the <b>conditional probability</b> of an event A given B.
* Conditional probability looks at the <b>probability of a second event occurring</b> based on the <b>probability of the first event occurring</b>.

<br>

### <font color='green'>Applications of Naive Bayes Classifier</font>

<b><font color='sky blue'>1) Spam filtering</font></b>
* Email services use this algorithm to figure out whether an email is a spam or not.

<b><font color='sky blue'>2) Sentiment Analysis</font></b>
* Sentiment Analysis refers to the identification of positive or negative sentiments of a target group (customers, audience, etc.)

<b><font color='sky blue'>3) Face recognition</font></b>
* Face recognition is a method of identifying or verifying the identity of an individual using their face. 

<br>

### <font color='green'>Naive Bayes Classifier equation</font>

* <b>P(A|B)</b> - Conditional probability of A given B
* <b>P(B|A)</b> - Conditional probability of B given A
* <b>P(A)</b> - Probability of event A
* <b>P(B)</b> - Probability of event B

<br>

### <font color='green'>Example for Naive Bayes Classifier</font>

* To predict whether a person will purchase a product on a specific combination of <b>Day</b>, <b>Discount</b> and <b>Free delivery</b>.
* Consider we have a small sample dataset of 30 rows.



* Based on this dataset, we will populate <b>frequency tables</b> and <b>likelihood tables</b> (the chance of something happening) for each attribute.





* Let us use these 3 likelihood tables to calculate <b>whether a customer will purchase a product or not</b> on a specific combination of <b>Day</b>, <b>Discount</b> and <b>Free delivery</b>.
* Here, let us take a combination of below A and B factors and calculate <b>Naive Bayes Classifier equation</b>,
* Where <b>B</b> equals
    * Day = <font color='red'>Holiday</font>
    * Discount = <font color='red'>Yes</font>
    * Free delivery = <font color='red'>Yes</font>
* <b>A</b> = <font color='red'>No buy</font>




* Here, let us take a combination of below A and B factors and calculate <b>Naive Bayes Classifier equation</b>,
* Where <b>B</b> equals
    * Day = <font color='red'>Holiday</font>
    * Discount = <font color='red'>Yes</font>
    * Free delivery = <font color='red'>Yes</font>
* <b>A</b> = <font color='red'>Yes buy</font>




* <b>Probability of purchase = <font color='blue'>0.986</font></b>
* <b>Probability of no purchase = <font color='blue'>0.178</font></b>



* As <b>84.71%</b> is greater than <b>15.29%</b>, we can conclude that an average <b>customer will buy</b> on a <b>Holiday</b> with <b>Discount</b> and <b>Free Delivery</b>.
