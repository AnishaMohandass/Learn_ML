# <font color='dark green'>Naive Bayes Classifier</font>

* Naive Bayes is a powerful algorithm that is used for <b>text data analysis</b>.
* Naive Bayes Classifier is based on <b>"Bayes" theorem</b> which gives the <b>conditional probability</b> of an event A given B.
* Conditional probability looks at the <b>probability of a second event occurring</b> based on the <b>probability of the first event occurring</b>.

<br>

### <font color='green'>Applications of Naive Bayes Classifier</font>

<b><font color='sky blue'>1) Spam filtering</font></b>
* Email services use this algorithm to figure out whether an email is a spam or not.

<b><font color='sky blue'>2) Sentiment Analysis</font></b>
* Sentiment Analysis refers to the identification of positive or negative sentiments of a target group (customers, audience, etc.)

<b><font color='sky blue'>3) Face recognition</font></b>
* Face recognition is a method of identifying or verifying the identity of an individual using their face. 

<br>

### <font color='green'>Naive Bayes Classifier equation</font>

* <b>P(A|B)</b> - Conditional probability of A given B
* <b>P(B|A)</b> - Conditional probability of B given A
* <b>P(A)</b> - Probability of event A
* <b>P(B)</b> - Probability of event B

<br>

### <font color='crimson'>Sample Project for Naive Bayes Classifier</font>

* To perform text classification of News headlines and classify news into diffrent topics for a News website.

```python
#Importing the libraries

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_20newsgroups

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
#To download the 20news dataset

data = fetch_20newsgroups()
print(data.target_names)
```

```python
#Defining all the categories

categories = data.target_names
```

```python
#Training the data on these categories

train = fetch_20newsgroups(subset='train', categories=categories)
print(len(train.data)) #no. of articles to train
print("------------------------------------------------------")
print(train.data[0]) #to view the first article in train set
```

```python
#Testing the data for these categories

test = fetch_20newsgroups(subset='test', categories=categories)
print(len(test.data)) #no. of articles to test
print("------------------------------------------------------")
print(test.data[0]) #to view the first article in test set
```

```python
#Creating a model based on Multinomial Naive Bayes

model = make_pipeline(TfidfVectorizer(), MultinomialNB())
```

### <font color='sky blue'>TfidfVectorizer - express the importance of a word to a document</font>

<b>Term Frequency (tf)</b>
* tf = no. of times a term appears in a document.

<b>Inverse Document Frequency (idf)</b>
* idf = it is a measure of how common or rare a term is across the entire corpus of documents.
* 0 = if the word is common and appears in many documents.
* 1 = if the word is rare.

<b>Term Frequency - Inverse Document Frequency (tf-idf)</b>
* tf-idf value of a term in a document is the product of its tf and idf.
* The higher is the value the more relevent the term is in that document.

### <font color='sky blue'>MultinomialNB</font>
* The multinomial Naive Bayes classifier is suitable for classification with discrete features (e.g., word counts for text classification).
* The program guesses the tag of a text, such as an email or a newspaper story, using the Bayes theorem. 

```python
#Training the model with the train data

model.fit(train.data, train.target)
```

```python
#To predict the values for the test data

predictions = model.predict(test.data)
```

```python
#Determining the accuracy of the model

accuracy = metrics.accuracy_score(test.target, predictions)
print(accuracy)

# That is, 77% Accuracy
```

```python
#Creating confusion matrix and heat map

mat = confusion_matrix(test.target, predictions)
```

```python
#Representing the confusion matrix in a heat map

fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(mat.T, annot=True, fmt='d', linewidths=.5, square=True, xticklabels=train.target_names, yticklabels=train.target_names, ax=ax)
plt.xlabel('True label')
plt.ylabel('Predicted label')
```

```python
#Predicting category on new data based on trained model

def predict_category(s, train=train, model=model):
    pred = model.predict([s])
    return train.target_names[pred[0]]
```

```python
predict_category("Rishi Sunak will be Britain's next prime minister")
```

```python
predict_category("Python is a powerful programming language.")
```

```python
predict_category("Angelina Jolie is an American actress, filmmaker, and humanitarian.")
```

```python
predict_category("For a family of 4-5, the top 3 most popular cars are Tata Altroz (10.25 Lakh onwards) and Maruti Suzuki Grand Vitara (10.45 Lakh onwards).")
```
