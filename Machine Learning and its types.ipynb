# <font color='orange'>Machine Learning</font>

* Humans can train the machines to learn from the past data and act much faster.

## <font color='orange'>Types of Machine Learning</font>

* There are 3 ways in which the machine learns,


## <font color='sky blue'>Supervised Learning</font>

* Supervised Learning uses labeled data to train the model.

<b><font color='brown'>For example:</font></b>

* Consider we have ₹1, ₹2 and ₹5 coins.



* Where the weight of ₹1 coin is 3g, ₹2 coin is 4g and ₹5 coin is 7g.
* Here, the weight of the coin is <b>feature</b> and the value of currency is <b>label</b>.
* The model is trained based on which feature is associated with which label.

## <font color='sky blue'>Unsupervised Learning</font>

* Unsupervised Learning uses unlabeled data to train the model.

<b><font color='brown'>For example:</font></b>

* Consider we have a cricketers data with their score and wickets.
* After plotting the score and wickets data, we get <b>Batsmen</b> and <b>Bowlers</b> (that is, labeled data).



## <font color='sky blue'>Reinforcement Learning</font>

* Reinforcement Learning works based on the feedback.

<b><font color='brown'>For example:</font></b>
    
* A dog image is given to the machine to predict which animal's image is it.
* The machine wrongly predicts the given image (that is, the machine says the given image is a cat).
* We will provide a feedback to the machine saying that it is a dog's image.
* The machine will learn based on the given feedback and predicts it correctly the next time when the dog's image is given.

## <font color='orange'>Methods to solve a problem in Machine Learning</font>

There are 3 methods to slove a problem in Machine Learning,

1) Classification

2) Regression

3) Clustering



<br>

## <font color='sky blue'>Classification</font>

* classification is a <b>supervised learning</b> technique.
* Classification algorithms look at existing data and predicts what a new data belongs to.

### <font color='olive'>Algorithms falls under Classification</font>

1) Decision Tree algorithm

2) Naive Bayes algorithm

3) Random Forest algorithm

4) Logistic Regression algorithm

5) K-nearest neighbors algorithm (KNN)

<br>

## <font color='sky blue'>Regression</font>

* Regression is a <b>supervised learning</b> technique.
* With a regression, no classified labels are predicted. It is used when the predicted data is numerical in nature.

### <font color='olive'>Algorithms falls under Regression</font>

Linear Regression algorithm

<br>

## <font color='sky blue'>Clustering</font>

* Clustering is a <b>unsupervised learning</b> technique.
* With clustering, the algorithm tries to find a pattern in data sets without labels associated with it. 

### <font color='olive'>Algorithms falls under Clustering</font>

1) K-Means algorithm
